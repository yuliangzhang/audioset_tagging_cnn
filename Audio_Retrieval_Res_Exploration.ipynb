```python
import pandas as pd
import numpy as np
```

# Load the first configure of clipwise retrieval ressults

```python
clipwise_res1 = pd.read_csv("results_bk/dcase_val_neighbors_topK_eval_bal_train_unbal.csv")
print("Shape size of clipwise res one:", clipwise_res1.shape)
clipwise_res1.head()
```

# Load the second configure of clipwise retrieval results 

```python
clipwise_res2 = pd.read_csv("results_bk_150/dcase_val_neighbors_topK150_min06_eval_bal_train_unbal.csv")
print("Shape size of clipwise res two:", clipwise_res2.shape)
clipwise_res2.head()
```

# Combine the two clipwise retrieval results and remove duplicated rows

```python
clipwise_res = pd.concat([clipwise_res1, clipwise_res2])
print("combined data size: ", clipwise_res.shape)
clipwise_res.drop_duplicates(subset=['query_segment_id', 'query_labels', 'candidate_video_id'],inplace=True)
print("remove duplicate samples after combine: ", clipwise_res.shape)
```

# Load segment-based retrieval results

```python
seg_wise_res = pd.read_csv("results2/dcase_val_window_neighbors_topK300_min055_full_merged.csv")
print("Shape size of segment-wise results: ", seg_wise_res.shape)
seg_wise_res.head()
```

## Investigate the number of audio clips for each class

```python
seg_wise_res.groupby("query_label", as_index=False)['candidate_video_id'].nunique().sort_values(by='query_label')
```

# Extract the audios that in both the clip-wise and segment-wise retrieval results

```python
clipwise_audio_list = set(clipwise_res['candidate_video_id'])
seg_wise_audio_list = set(seg_wise_res['candidate_video_id'])

print("clip-wise audio list size is ", len(clipwise_audio_list))
print("segment-wise audio list size is ", len(seg_wise_audio_list))

intersect_audio_list = clipwise_audio_list.intersection(seg_wise_audio_list)
print("The intersection of clip-wise and segment-wise audio list size is ", len(intersect_audio_list))
```

## Check the statistics of the intersected audio clips for each class

```python
intersect_seg_wise_res = seg_wise_res[seg_wise_res['candidate_video_id'].isin(intersect_audio_list)]
print("The original size of intersected segment-wise results is ", intersect_seg_wise_res.shape)

intersect_seg_wise_res.groupby("query_label", as_index=False)['candidate_video_id'].nunique().sort_values(by='query_label')
```

# Select high similarity samples from clip-wise and segment-wise results separately and combine them together

```python
# clipwise_high_audio_list = set(clipwise_res[(clipwise_res['similarity'] > 0.7) & (~(clipwise_res['query_labels'].str.contains('Speech')))]['candidate_video_id'])
# segwise_high_audio_list = set(seg_wise_res[(seg_wise_res['similarity'] > 0.7) & (~(seg_wise_res['query_label'].str.contains('Speech')))]['candidate_video_id'])
clipwise_res_filt = clipwise_res.dropna(subset=['query_labels'])
clipwise_cond = ~clipwise_res_filt['query_labels'].str.contains('Speech') & (clipwise_res_filt['similarity'] > 0.7)

seg_wise_res_filt = seg_wise_res.dropna(subset=['query_label'])
segwise_cond = ~seg_wise_res_filt['query_label'].str.contains('Speech') & (seg_wise_res_filt['similarity'] > 0.7)
clipwise_high_audio_list = set(clipwise_res_filt[clipwise_cond]['candidate_video_id'])
segwise_high_audio_list = set(seg_wise_res_filt[segwise_cond]['candidate_video_id'])

print('clip-wise high similarity audio number is ', len(clipwise_high_audio_list))
print('segment-wise high similarity audio number is ', len(segwise_high_audio_list))
```

```python
final_selected_audio_list = intersect_audio_list.union(clipwise_high_audio_list).union(segwise_high_audio_list)
print("intersect: ", len(intersect_audio_list))
print("clipwise high: " , len(clipwise_high_audio_list))
print("segwise high: ", len(segwise_high_audio_list))
print("final selected: ", len(final_selected_audio_list))
```

```python
clipwise_res.columns
```

```python

clip_selected_res = clipwise_res[clipwise_res['candidate_video_id'].isin(final_selected_audio_list)][['query_segment_id', 'query_labels', 'candidate_segment_id',
       'candidate_video_id', 'similarity', "candidate_label_indices",	'candidate_label_names', 'source_split']]
clip_selected_res = clip_selected_res.rename(columns={"query_labels": "query_label"})

clip_selected_res.head()
```

```python
seg_selected_res = seg_wise_res[seg_wise_res['candidate_video_id'].isin(final_selected_audio_list)][['query_segment_id', 'query_label', 'candidate_segment_id',
       'candidate_video_id', 'similarity', "candidate_label_indices",	'candidate_label_names', 'source_split']]
seg_selected_res.head()
```

```python
final_selected_res = pd.concat([seg_selected_res, clip_selected_res])
final_selected_res = final_selected_res.loc[final_selected_res.groupby(['query_segment_id', 'query_label', 'candidate_segment_id',
       'candidate_video_id'])['similarity'].idxmax()].reset_index(drop=True)

print(final_selected_res.shape)
```

```python
len(set(final_selected_res['candidate_segment_id']))
```

# Analyze the final selected audios

```python
for thresh in np.arange(0,1, 0.1):
    tmp_res = final_selected_res[final_selected_res['similarity'] > thresh]
    print("threshold: ", thresh, " Number of audio kept: ", len(set(tmp_res['candidate_segment_id'])))


final_selected_res = final_selected_res[final_selected_res['similarity'] > 0.65]
```
